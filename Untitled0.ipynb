```python
SevenSigma(df)
df.drop(columns=['max_glu_serum','A1Cresult','weight','citoglipton','examide','payer_code','medical_specialty'],inplace=True)
print("Value counts for 'readmitted' after initial column drops:")
print(df['readmitted'].value_counts())
df['race'].value_counts()
```

```python
df['race'] = df['race'].apply(lambda x: 'Other' if x == '?' else x)

# Robustly drop specific indices
indices_to_drop = [30506, 75551, 82573]
existing_indices_to_drop = [idx for idx in indices_to_drop if idx in df.index]
if existing_indices_to_drop:
    df.drop(index=existing_indices_to_drop, inplace=True)
else:
    print(f"Warning: None of the specified indices {indices_to_drop} found in DataFrame for dropping.")

index=[]
index=list(df[df['diag_1']=='?'].index)
index.extend(df[df['diag_2']=='?'].index)
index.extend(df[df['diag_3']=='?'].index)

# Convert to set to remove duplicates, then back to list to ensure unique indices are dropped
index_to_drop_diag = list(set(index))
existing_diag_indices_to_drop = [idx for idx in index_to_drop_diag if idx in df.index]
if existing_diag_indices_to_drop:
    df.drop(index=existing_diag_indices_to_drop, inplace=True)
else:
    print("Warning: No '?' values found in diag columns for dropping.")

def replace_age_ranges_correctly(series):
    age_map = {
        '[0-10)': 1,
        '[10-20)': 2,
        '[20-30)': 3,
        '[30-40)': 4,
        '[40-50)': 5,
        '[50-60)': 6,
        '[60-70)': 7,
        '[70-80)': 8,
        '[80-90)': 9,
        '[90-100)': 10
    }
    return series.map(age_map)

print("df['age'] dtype BEFORE age mapping:", df['age'].dtype)
print("df['age'] value_counts() BEFORE age mapping:")
print(df['age'].value_counts())
df['age'] = replace_age_ranges_correctly(df['age'])
print("df['age'] dtype AFTER age mapping:", df['age'].dtype)
print("df['age'] value_counts() AFTER age mapping:")
print(df['age'].value_counts())

df['admission_type_id']=df['admission_type_id'].apply(lambda x : 5 if x in (6,8) else x)
df['admission_type_id']=df['admission_type_id'].apply(lambda x : 1 if x == 4 else 2 if x==7 else x )
```

```python
df['discharge_disposition_id']=df['discharge_disposition_id'].apply(lambda x : 1 if x in (6,8) else x)

#Uncategorized/Unknown: 18, 25, 26, 12
df['discharge_disposition_id']=df['discharge_disposition_id'].apply(lambda x : 18 if x in (25,26,12) else x)

#Expired:11, 19, 20, 21
df['discharge_disposition_id']=df['discharge_disposition_id'].apply(lambda x : 11 if x in (19,20,21) else x)

#Hospice:13, 14
df['discharge_disposition_id']=df['discharge_disposition_id'].apply(lambda x : 13 if x ==14 else x)

# Discharged/Transferred to Hospital: 2, 9, 10, 23, 27, 28, 29
df['discharge_disposition_id']=df['discharge_disposition_id'].apply(lambda x : 2 if x in ( 9, 10, 23, 27, 28, 29) else x)
df['discharge_disposition_id']=df['discharge_disposition_id'].apply(lambda x : 3 if x in ( 4, 5, 15, 24) else x)

#Discharged to Outpatient Services:16, 17
df['discharge_disposition_id']=df['discharge_disposition_id'].apply(lambda x : 16 if x ==17 else x)
# Unknown/Invalid: 9, 15, 17, 20, 21
df['admission_source_id']= df['admission_source_id'].apply(lambda x : 9 if x in (15, 17, 20, 21) else x)

# Physician/Clinic Referral:1, 2, 3
df['admission_source_id']= df['admission_source_id'].apply(lambda x : 1 if x in (2,3) else x)
df['admission_source_id']= df['admission_source_id'].apply(lambda x : 4 if x in (10,22) else x)


#Transfer from Facility:5, 6, 18, 19, 25, 26
df['admission_source_id']= df['admission_source_id'].apply(lambda x : 5 if x in (6, 18, 19, 25, 26) else x)

#Delivery:11, 12, 13, 14
df['admission_source_id']= df['admission_source_id'].apply(lambda x : 11 if x in (12, 13, 14) else x)
# all three diag features have more than 700 unqie we are grouping them based on ICD-9 codes
diag1_list=[]
diag2_list=[]
diag3_list=[]
def diag_cluster(col, diag_list):
    for x in df[col]:
        x_val = str(x) # Convert x to string to avoid TypeError if x is an int
        # If the value in the 'col' column contains 'V' or 'E', it is assigned a cluster value of 18.
        if 'V' in x_val or 'E' in x_val:
            diag_list.append(18)
            continue
        # The following conditions assign cluster values based on specific ranges of float values.
        try:
            x_float = float(x_val)
            if 1 <= x_float <= 139:
                diag_list.append(1)
            elif 140 <= x_float <= 239:
                diag_list.append(2)
            elif 240 <= x_float <= 279:
                diag_list.append(3)
            elif 280 <= x_float <= 289:
                diag_list.append(4)
            elif 290 <= x_float <= 319:
                diag_list.append(5)
            elif 320 <= x_float <= 389:
                diag_list.append(6)
            elif 390 <= x_float <= 459:
                diag_list.append(7)
            elif 460 <= x_float <= 519:
                diag_list.append(8)
            elif 520 <= x_float <= 579:
                diag_list.append(9)
            elif 580 <= x_float <= 629:
                diag_list.append(10)
            elif 630 <= x_float <= 679:
                diag_list.append(11)
            elif 680 <= x_float <= 709:
                diag_list.append(12)
            elif 710 <= x_float <= 739:
                diag_list.append(13)
            elif 740 <= x_float <= 759:
                diag_list.append(14)
            elif 760 <= x_float <= 779:
                diag_list.append(15)
            elif 780 <= x_float <= 799:
                diag_list.append(16)
            elif 800 <= x_float <= 999:
                diag_list.append(17)
        except ValueError:
            # If it's not 'V'/'E' and also not a valid float (e.g., unexpected non-numeric string)
            diag_list.append(18) # Assign to 'other' category
diag_cluster('diag_1',diag1_list)
diag_cluster('diag_2',diag2_list)
diag_cluster('diag_3',diag3_list)
df['diag_1']=diag1_list
df['diag_2']=diag2_list
df['diag_3']=diag3_list
df['change']=df['change'].apply(lambda x: 'Yes' if x=='Ch' else x)
medicines = ['metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide', 'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone', 'tolazamide', 'insulin', 'glyburide-metformin', 'glipizide-metformin', 'glimepiride-pioglitazone', 'metformin-rosiglitazone', 'metformin-pioglitazone']

for col in medicines:
    df[col] = df[col].apply(lambda x: 0 if x == 'No' else 1)
df['gender']=df['gender'].apply(lambda x: 0 if x=='Female' else 1)
df['change']=df['change'].apply(lambda x : 0 if x=='No' else 1)
df['diabetesMed']=df['diabetesMed'].apply(lambda x : 0 if x=='No' else 1)

print("Value counts for 'readmitted' before transformation:")
print(df['readmitted'].value_counts())
df['readmitted']=df['readmitted'].apply(lambda x : 1 if x=='<30' else 0)
print("Value counts for 'readmitted' after transformation:")
print(df['readmitted'].value_counts())

# Robustly handle drop_duplicates and drop columns
if 'patient_nbr' in df.columns:
    print("Value counts for 'readmitted' before drop_duplicates:")
    print(df['readmitted'].value_counts())
    df.drop_duplicates(subset='patient_nbr',keep='first',inplace=True)
    print("Value counts for 'readmitted' after drop_duplicates:")
    print(df['readmitted'].value_counts())
else:
    print("Warning: 'patient_nbr' not found for drop_duplicates. Skipping.")

columns_to_drop = []
if 'encounter_id' in df.columns:
    columns_to_drop.append('encounter_id')
if 'patient_nbr' in df.columns:
    columns_to_drop.append('patient_nbr')

if columns_to_drop:
    df.drop(columns=columns_to_drop, inplace=True)
else:
    print("Warning: 'encounter_id' and 'patient_nbr' not found for dropping. Skipping.")

# used label encoder for race feature.
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['race']=encoder.fit_transform(df['race'])
df.shape
X=df[['num_lab_procedures', 'num_medications', 'diag_3', 'diag_1', 'age',
       'diag_2', 'time_in_hospital', 'number_diagnoses', 'num_procedures',
       'admission_source_id', 'race', 'admission_type_id',
       'discharge_disposition_id', 'insulin', 'diabetesMed', 'change',
       'metformin', 'glipizide', 'number_outpatient', 'glyburide',
       'rosiglitazone', 'pioglitazone', 'number_inpatient', 'glimepiride',
       'number_emergency']]
y=df['readmitted']

# Check distribution of 'readmitted' before SMOTEN
print("Value counts for 'readmitted' column before SMOTEN (final check):")
print(y.value_counts())

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
from imblearn.over_sampling import SMOTEN

# Check if y_train has more than one class before applying SMOTEN
if len(y_train.unique()) > 1:
    sm=SMOTEN(sampling_strategy='minority',random_state=42)
    X_res, y_res = sm.fit_resample(X_train, y_train)
else:
    print("Error: y_train has only one class, SMOTEN cannot be applied.")
    print("Please check the preprocessing steps as the minority class appears to be missing or eliminated.")
```

```python
cls = RandomForestClassifier(n_estimators=100, random_state=42)
cls.fit(X_res, y_res)
cls_pred = cls.predict(X_test)
print(classification_report(y_test, cls_pred))
```

```python
import tensorflow as tf

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(32, activation = 'relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(16, activation = 'relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(8, activation = 'relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(4, activation = 'relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
model.compile(tf.keras.optimizers.Adam(learning_rate = 0.0001), tf.keras.losses.BinaryCrossentropy(), metrics=['acc'])
model.fit(X_res.to_numpy(), y_res, batch_size=128, epochs=400, validation_split=0.25)
```

```python
model_pred = model.predict(X_test.to_numpy()) > 0.5
print(classification_report(y_test, model_pred))
```
